```
from newspaper import Article
```

```
foxnews_articles = []
foxnews_text = []
for url in data:
    article = Article(url.get('url'))
    article.download()
    article.parse()
    foxnews_articles.append(article)
    foxnews_text.append(str(article.text))
```

```
with open('foxnews.txt', 'w') as filehandle:
    for text in foxnews_text:
        filehandle.write('%s\n' % text)
filehandle.close()
```

```
with open("/content/RISJbot/jsonloutput/abc/2021-02-26T14-43-20-abc.jsonl", "r") as read_file:
    abc_data = [json.loads(line) for line in read_file]
```

```
abc_articles = []
abc_text = []
for url in abc_data:
    article = Article(url.get('url'))
    article.download()
    article.parse()
    abc_articles.append(article)
    abc_text.append(str(article.text))
```

```
with open('abc.txt', 'w') as filehandle:
    for text in abc_text:
        filehandle.write('%s\n' % text)
filehandle.close()
```

```
import spacy
!python -m spacy download en_core_web_sm
```

```
nlp = spacy.load('en_core_web_sm')
```

```
abc_tokens = []
for text in abc_text:
    for token in nlp(text):
        if not token.is_stop and not token.is_punct:
            abc_tokens.append(str(token.lemma_))

print(len(abc_tokens))
```

```
fox_tokens = []
for text in foxnews_text:
    for token in nlp(text):
        if not token.is_stop and not token.is_punct:
            fox_tokens.append(str(token.lemma_))

print(len(fox_tokens))
```

```
print(len(foxnews_text))
print(len(abc_text))
```

```
import numpy
fox_np = np.asarray(fox_tokens)
abc_np = np.asarray(abc_tokens)
vocab = np.concatenate((fox_np,abc_np))
print(len(vocab))
print(len(np.unique(vocab)))
```

```
i = 0
types = set()
type_counts = []
for token in fox_np:
    if token not in types:
        types.add(token)
        i += 1
    type_counts.append(i)
```

```
from matplotlib import pyplot as plt
plt.plot(list(range(len(fox_np))),type_counts)
plt.xlabel('#Tokens')
plt.ylabel('#Types')
plt.ylim(top=len(fox_np))
plt.show()
```

```
i = 0
types = set()
type_counts = []
for token in abc_np:
    if token not in types:
        types.add(token)
        i += 1
    type_counts.append(i)
```

```
from matplotlib import pyplot as plt
plt.plot(list(range(len(abc_np))),type_counts)
plt.xlabel('#Tokens')
plt.ylabel('#Types')
plt.ylim(top=len(abc_np))
plt.show()
```
